# 1. データ生成

```python
# 利用するライブラリのインポート
import numpy as np
import pandas as pd
import statsmodels.api as sm

# 共通のパラメータ
alpha = 10
beta = 5
n_samples = 1000

# Scenario 1: X and u are correlated
np.random.seed(42)
X_corr = np.random.rand(n_samples) * 20
u_corr = X_corr* 1+ np.random.randn(n_samples) * 5 # Introduce correlation
Y_corr = alpha + beta * X_corr+ u_corr
data_corr = pd.DataFrame({'Y': Y_corr, 'X': X_corr})

# Scenario 2: X and u are uncorrelated
np.random.seed(42)
X_uncorr = np.random.rand(n_samples)* 20 # Education years, 0-20
u_uncorr = np.random.randn(n_samples)* 5 # Other factors
Y_uncorr = alpha + beta * X_uncorr+ u_uncorr
data_uncorr = pd.DataFrame({'Y': Y_uncorr, 'X': X_uncorr})

print("Scenario 1: Corr", np.corrcoef(X_corr,u_corr)[0, 1])
print("Scenario 2: Corr", np.corrcoef(X_uncorr,u_uncorr)[0, 1])
```

# 2. 相関ありのデータで回帰分析

```python
# 相関ありのデータで回帰分析
X_corr_const = sm.add_constant(data_corr['X'])
model_corr = sm.OLS(data_corr['Y'], X_corr_const).fit()
print("---OLS Results (Correlated X and u) ---")
print(model_corr.summary())
```

# 3. 無相関データで回帰分析

```python
# 相関なしのデータで回帰分析
X_uncorr_const = sm.add_constant(data_uncorr['X'])
model_uncorr = sm.OLS(data_uncorr['Y'], X_uncorr_const).fit()
print("---OLS Results (Uncorrelated X and u) ---")
print(model_uncorr.summary())
```
